```python
# Numerical computations
import numpy as np

# Deep Learning Framework
import tensorflow as tf

# Text preprocessing
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Neural Network Layers
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense

# Build the complete model
from tensorflow.keras.models import Model
```

```python
print("TensorFlow Version:", tf.__version__)
print("NumPy Version:", np.__version__)
```

```python
english_sentences = [
    "hello",
    "hi",
    "good morning",
    "good night",
    "goodbye",
    "thank you",
    "sorry",
    "please",
    "yes",
    "no",
    "how are you",
    "i am fine",
    "what is your name",
    "my name is debasish",
    "where are you",
    "come here",
    "sit down",
    "stand up",
    "open the door",
    "close the door"
]

bengali_sentences = [
    "а¦№а§Ќа¦Їа¦ѕа¦Іа§‹",
    "а¦№а¦ѕа¦‡",
    "а¦ёа§Ѓа¦Єа§Ќа¦°а¦­а¦ѕа¦¤",
    "а¦¶а§Ѓа¦­ а¦°а¦ѕа¦¤а§Ќа¦°а¦ї",
    "а¦¬а¦їа¦¦а¦ѕа¦Їа¦ј",
    "а¦§а¦Ёа§Ќа¦Їа¦¬а¦ѕа¦¦",
    "а¦¦а§Ѓа¦ѓа¦–а¦їа¦¤",
    "а¦…а¦Ёа§Ѓа¦—а§Ќа¦°а¦№ а¦•а¦°а§‡",
    "а¦№а§Ќа¦Їа¦ѕа¦Ѓ",
    "а¦Ёа¦ѕ",
    "а¦¤а§Ѓа¦®а¦ї а¦•а§‡а¦®а¦Ё а¦†а¦›а§‹",
    "а¦†а¦®а¦ї а¦­а¦ѕа¦Іа§‹ а¦†а¦›а¦ї",
    "а¦¤а§‹а¦®а¦ѕа¦° а¦Ёа¦ѕа¦® а¦•а§Ђ",
    "а¦†а¦®а¦ѕа¦° а¦Ёа¦ѕа¦® а¦¦а§‡а¦¬а¦ѕа¦¶а§Ђа¦·",
    "а¦¤а§Ѓа¦®а¦ї а¦•а§‹а¦Ґа¦ѕа¦Їа¦ј а¦†а¦›а§‹",
    "а¦Џа¦–а¦ѕа¦Ёа§‡ а¦†а¦ёа§‹",
    "а¦¬а¦ёа§‹",
    "а¦¦а¦ѕа¦Ѓа¦Ўа¦ја¦ѕа¦“",
    "а¦¦а¦°а¦ња¦ѕ а¦–а§‹а¦Іа§‹",
    "а¦¦а¦°а¦ња¦ѕ а¦¬а¦Ёа§Ќа¦§ а¦•а¦°а§‹"
]
```

```python
decoder_bengali_sentences=[]

for words in bengali_sentences:
  decoder_bengali_sentences.append("<start> "+words+" <end>")

print(decoder_bengali_sentences)
```

```python
english_tokenizer=Tokenizer()
bengali_tokenizer = Tokenizer(
    filters='!"#$%&()*+,-./:;=?@[\\]^_`{|}~\t\n'
)
english_tokenizer.fit_on_texts(english_sentences)
bengali_tokenizer.fit_on_texts(decoder_bengali_sentences)

print(english_tokenizer.word_index)
print(bengali_tokenizer.word_index)
```

```python
english_sequences = english_tokenizer.texts_to_sequences(
    english_sentences
)

bengali_sentences=decoder_bengali_sentences

bengali_sequences = bengali_tokenizer.texts_to_sequences(
    bengali_sentences
)

print(english_sequences)
print(bengali_sequences)
```

```python
 maximum_eng_length=0
 maximum_ben_length=0

for seq in english_sequences:
  maximum_eng_length=max(maximum_eng_length,len(seq))
print(maximum_eng_length)

for seq in bengali_sequences:
  maximum_ben_length=max(maximum_ben_length,len(seq))
print(maximum_ben_length)
```

```python
english_padded=pad_sequences(
    english_sequences,
    maxlen = maximum_eng_length,
    padding ='post'
)

bengali_padded=pad_sequences(
    bengali_sequences,
    maxlen = maximum_ben_length,
    padding ='post'
)

print(english_padded)
print(bengali_padded)
```

```python
english_vocab_size = len(english_tokenizer.word_index) + 1
bengali_vocab_size = len(bengali_tokenizer.word_index) + 1

```

```python
encoder_inputs = Input(shape=(maximum_eng_length,))

encoder_embedding = Embedding(
    input_dim = english_vocab_size,
    output_dim = 16,
)(encoder_inputs)

encode_lstm=LSTM(16,return_state=True)

encoder_outputs,state_h,state_c=encode_lstm(encoder_embedding)

encoder_states=[state_h,state_c]
```

```python
decoder_inputs=Input(shape=(maximum_ben_length-1,))

decoder_embedding_layer=Embedding(
    input_dim=bengali_vocab_size,
    output_dim=16)

decoder_lstm=LSTM(16,return_sequences=True,return_state=True)

decoder_outputs,_,_=decoder_lstm(decoder_embedding_layer(decoder_inputs),initial_state=encoder_states
)

decoder_dense=Dense(bengali_vocab_size,activation='softmax')

decoder_outputs=decoder_dense(decoder_outputs)
```

```python
model = Model(
    inputs=[encoder_inputs, decoder_inputs],
    outputs=decoder_outputs
)
```

```python
encoder_input_data = english_padded

decoder_input_data=bengali_padded[:, :-1]
decoder_target_data=bengali_padded[:,1:]

print(decoder_input_data)
print("----------------------------------------")
print(decoder_target_data)
```

```python
model = Model(
    inputs=[encoder_inputs, decoder_inputs],
    outputs=decoder_outputs
)

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    epochs=300,
    batch_size=4,
    verbose=1
)
```

```python
encoder_model = Model(
    inputs=encoder_inputs,
    outputs=encoder_states
)
```

```python
decoder_state_input_h = Input(shape=(16,))
decoder_state_input_c = Input(shape=(16,))

decoder_states_inputs = [
    decoder_state_input_h,
    decoder_state_input_c
]
```

```python
decoder_embedding_inf = decoder_embedding_layer(decoder_inputs)

decoder_outputs, state_h, state_c = decoder_lstm(
    decoder_embedding_inf,
    initial_state=decoder_states_inputs
)

decoder_states = [state_h, state_c]

decoder_outputs = decoder_dense(decoder_outputs)
```

```python
decoder_model = Model(
    inputs=[decoder_inputs] + decoder_states_inputs,
    outputs=[decoder_outputs] + decoder_states
)
```

```python
reverse_bengali = {
    index: word
    for word, index in bengali_tokenizer.word_index.items()
}
```

```python
import numpy as np

def translate(sentence):

    # ---------- Encoder ----------

    sequence = english_tokenizer.texts_to_sequences([sentence])

    sequence = pad_sequences(
        sequence,
        maxlen=maximum_eng_length,
        padding="post"
    )

    states = encoder_model.predict(sequence, verbose=0)

    # ---------- Decoder ----------

    target_seq = np.array(
        [[bengali_tokenizer.word_index["<start>"]]]
    )

    translated_sentence = []

    while True:

        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states,
            verbose=0
        )

        sampled_token = np.argmax(
            output_tokens[0, -1, :]
        )

        sampled_word = reverse_bengali.get(
            sampled_token,
            ""
        )

        if sampled_word == "<end>":
            break

        translated_sentence.append(sampled_word)

        target_seq = np.array([[sampled_token]])

        states = [h, c]

        if len(translated_sentence) > maximum_ben_length:
            break

    return " ".join(translated_sentence)
```

```python
print(translate("hello"))
print(translate("good morning"))
print(translate("thank you"))
print(translate("i am fine"))
print(translate("where are you"))
```
